```python
###GRID BASED ON SELF DEFINED LINES AND DIMENSTIONS


#define different regions
sipm_ring = plt.Circle((0, 0), radius = 14.45, linewidth = 3, linestyle = '-',color='m', fill=False)


#draw mesh lines
wire_thickness = 0.150 #in mm, 150 um
wire_pitch = 2.45 #in mm
tpc_diameter = 45 #in mm
offset_plus = 1
x = np.arange(-22.5,22.5,wire_pitch)

#make list of positive y values
y_plus = []
for lp in range(18): #loop over the number of lines we need
    line_plus = []
    for p in range(len(x)): #loop over all the x values to make y values
        line_plus.append(0.5*x[p] + lp*wire_pitch + offset_plus) #y values for single line scaled by line number
    y_plus.append(line_plus)
               
#make list of negative y values
y_minus = []
for lm in range(18): #loop over the number of lines we need
    line_minus = []
    for p in range(len(x)): #loop over all the x values to make y values
        line_minus.append(0.5*x[p] - lm*wire_pitch + offset_plus) #y values for single line scaled by line number
    y_minus.append(line_minus)

fig, ax = plt.subplots()
h = ax.hist2d(xvals_190604_161603,yvals_190604_161603, bins=100,range = ((-25,25),(-25,25)), norm=LogNorm())

#plot mesh wires...
for wire in y_plus:
    plt.plot(x,wire,color ='k',linestyle='-')
    
for wire in y_minus:
    plt.plot(x,wire,color ='k',linestyle='-')

plt.title('Distribution of Reconstructed S2s')
plt.xlabel('x (mm)')
plt.ylabel('y (mm)')
ax.add_artist(sipm_ring)
ax.set_xlim((-25, 25))
ax.set_ylim((-25, 25))
ax.set_aspect(1)
plt.colorbar(h[3], ax=ax)

# plt.plot(x,y,linestyle = '-',color = 'k')
# plt.plot(x,y1,linestyle = '-',color = 'k')

#draw anode mesh

# plt.xlim(-5,5)

# plt.rc('grid', linestyle="-", color='black')
# plt.rcParams['lines.linewidth'] = 1
# plt.grid(True)
# fig.savefig('./posrec_figs/reconstruction_190604_161603.pdf')
plt.show()
```
